```python
import pandas as pd
```

```python
df = pd.read_csv('Traffic_Crashes.csv', parse_dates=['CRASH_DATE'], date_format='%m/%d/%Y %I:%M:%S %p')
```

```python
df = df.drop(columns=['CRASH_RECORD_ID', 'CRASH_DATE_EST_I'])
```

```python
df.head()
```

Number of crashes by year

```python
df['CRASH_YEAR'] = df['CRASH_DATE'].dt.year
crashes_by_year = df['CRASH_YEAR'].value_counts()
print(crashes_by_year)
```

```python
crashes_by_year
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=crashes_by_year.index, y=crashes_by_year.values, order=crashes_by_year.index.sort_values(), color='green')
plt.xlabel("Year")
plt.ylabel("Number of Crashes")
plt.title("Number of Crashes by Year")
plt.xticks(rotation=45)
plt.show()
```

**Get total sum from the crashes_by_year**

```python
total_crashes = crashes_by_year.sum()
print(f"Total number of crashes: {total_crashes}")
```

```python
df
```

```python
crashes_by_year_weather = df.groupby(['CRASH_YEAR', 'WEATHER_CONDITION']).size()
print(crashes_by_year_weather)
```

```python
crashes_by_year_weather
```

```python
crashes_by_year_weather_pivot = crashes_by_year_weather.unstack()
display(crashes_by_year_weather_pivot)
```

```python
crashes_by_year_weather_pivot.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.xlabel("Year")
plt.ylabel("Number of Crashes")
plt.title("Number of Crashes by Year and Weather Condition")
plt.legend(title='Weather Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
